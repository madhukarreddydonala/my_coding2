```python
import torch
from transformers import AutoModelForSequenceClassification, AutoTokenizer
import requests
from bs4 import BeautifulSoup
import re
model_name="cardiffnlp/tweet-topic-21-multi"
model=AutoModelForSequenceClassification.from_pretrained(model_name)
tokenizer=AutoTokenizer.from_pretrained(model_name)
labels=[
    "arts_&_culture","business_&_entrepreneurs","celebrity_&_pop_culture","diaries_&_daily_life",
    "family","fashion_&_style","film"
]
```

```python
texts=[
    "the last iphone was just released with an incredible new camera"
    "Manchester United won their match with a stunning goal in the last minute."
    "NASA just lancuhed a new mission to explore the surface of mars."
]
```

```python
inputs = tokenizer(texts, padding=True, truncation=True, return_tensors="pt")
with torch.no_grad():
    outputs = model(**inputs)
probabilities = torch.nn.functional.softmax(outputs.logits, dim=-1)
predictions = torch.argmax(probabilities, dim=-1)
```

```python
inputs = tokenizer(texts, padding=True, truncation=True, return_tensors="pt")
with torch.no_grad():
    outputs = model(**inputs)
probabilities = torch.nn.functional.softmax(outputs.logits, dim=-1)
predictions = torch.argmax(probabilities, dim=-1)
for text, pred,prob in zip(texts, predictions,probabilities):
    print(f"Text: {text}\nPredicted Label: {labels[pred.item()]}, confidence: {prob[pred].item():.4f} \n")

    # print(f"Text: {text}\nPredicted Label: {labels[pred.item()]}, confidence: {prob[pred].item():.4f} \n")
```

```python
from transformers import pipeline
summarizer = pipeline("summarization")
text="""
he user can navigate forwards and backwards through all
execution steps, and the visualization changes to match the
run-time state of the stack and heap at each step. In this example,
the user would see their custom LinkedList data
"""
```

```python
summary=summarizer(text,max_length=50,min_length=10,do_sample=False)
print(summary[0]['summary_text'])
```

```python
from transformers import AutoTokenizer, AutoModelForCausalLM
model_name = "gpt2"

tokenizer = AutoTokenizer.from_pretrained(model_name)

model = AutoModelForCausalLM.from_pretrained(model_name)
#
prompt=' once upon a time in a distant galaxy,'
input_ids = tokenizer.encode(prompt, return_tensors="pt")
output = model.generate(input_ids, max_length=100, num_return_sequences=1, no_repeat_ngram_size=2)
generated_text = tokenizer.decode(output[0], skip_special_tokens=True)
print(generated_text)
```
